```python
print("\n" + "="*80)
print("3. UNIVARIATE ANALYSIS")
print("="*80)

# 3.1 Distribution of ESG Scores with KDE and Statistics
fig, axes = plt.subplots(2, 2, figsize=(16, 12))
fig.suptitle('Distribution of ESG Scores (Histogram + KDE)', fontsize=18, fontweight='bold')

# ESG Overall
axes[0,0].hist(df['ESG_Overall'].dropna(), bins=35, color='#10B981', alpha=0.7, edgecolor='black', density=True)
df['ESG_Overall'].plot(kind='kde', ax=axes[0,0], color='darkgreen', linewidth=3)
axes[0,0].set_title('ESG Overall Score Distribution', fontsize=14, fontweight='bold')
axes[0,0].set_xlabel('ESG Overall Score', fontsize=12)
axes[0,0].set_ylabel('Density', fontsize=12)
axes[0,0].axvline(df['ESG_Overall'].mean(), color='red', linestyle='--', linewidth=2, 
                  label=f'Mean: {df["ESG_Overall"].mean():.2f}')
axes[0,0].axvline(df['ESG_Overall'].median(), color='orange', linestyle='--', linewidth=2, 
                  label=f'Median: {df["ESG_Overall"].median():.2f}')
axes[0,0].legend(fontsize=10)
axes[0,0].grid(alpha=0.3)

# ESG Environmental
axes[0,1].hist(df['ESG_Environmental'].dropna(), bins=35, color='#059669', alpha=0.7, edgecolor='black', density=True)
df['ESG_Environmental'].plot(kind='kde', ax=axes[0,1], color='darkslategray', linewidth=3)
axes[0,1].set_title('ESG Environmental Score Distribution', fontsize=14, fontweight='bold')
axes[0,1].set_xlabel('Environmental Score', fontsize=12)
axes[0,1].set_ylabel('Density', fontsize=12)
axes[0,1].axvline(df['ESG_Environmental'].mean(), color='red', linestyle='--', linewidth=2, 
                  label=f'Mean: {df["ESG_Environmental"].mean():.2f}')
axes[0,1].axvline(df['ESG_Environmental'].median(), color='orange', linestyle='--', linewidth=2, 
                  label=f'Median: {df["ESG_Environmental"].median():.2f}')
axes[0,1].legend(fontsize=10)
axes[0,1].grid(alpha=0.3)

# ESG Social
axes[1,0].hist(df['ESG_Social'].dropna(), bins=35, color='#FC8181', alpha=0.7, edgecolor='black', density=True)
df['ESG_Social'].plot(kind='kde', ax=axes[1,0], color='darkred', linewidth=3)
axes[1,0].set_title('ESG Social Score Distribution', fontsize=14, fontweight='bold')
axes[1,0].set_xlabel('Social Score', fontsize=12)
axes[1,0].set_ylabel('Density', fontsize=12)
axes[1,0].axvline(df['ESG_Social'].mean(), color='red', linestyle='--', linewidth=2, 
                  label=f'Mean: {df["ESG_Social"].mean():.2f}')
axes[1,0].axvline(df['ESG_Social'].median(), color='orange', linestyle='--', linewidth=2, 
                  label=f'Median: {df["ESG_Social"].median():.2f}')
axes[1,0].legend(fontsize=10)
axes[1,0].grid(alpha=0.3)

# ESG Governance
axes[1,1].hist(df['ESG_Governance'].dropna(), bins=35, color='#F59E0B', alpha=0.7, edgecolor='black', density=True)
df['ESG_Governance'].plot(kind='kde', ax=axes[1,1], color='darkorange', linewidth=3)
axes[1,1].set_title('ESG Governance Score Distribution', fontsize=14, fontweight='bold')
axes[1,1].set_xlabel('Governance Score', fontsize=12)
axes[1,1].set_ylabel('Density', fontsize=12)
axes[1,1].axvline(df['ESG_Governance'].mean(), color='red', linestyle='--', linewidth=2, 
                  label=f'Mean: {df["ESG_Governance"].mean():.2f}')
axes[1,1].axvline(df['ESG_Governance'].median(), color='orange', linestyle='--', linewidth=2, 
                  label=f'Median: {df["ESG_Governance"].median():.2f}')
axes[1,1].legend(fontsize=10)
axes[1,1].grid(alpha=0.3)

plt.tight_layout()
plt.savefig('../output/esg_distributions.png', dpi=300, bbox_inches='tight')
plt.show()

# 3.2 Financial Metrics Distribution with Statistics
fig, axes = plt.subplots(2, 2, figsize=(16, 12))
fig.suptitle('Distribution of Financial Metrics (Histogram + Statistics)', fontsize=18, fontweight='bold')

# Revenue
axes[0,0].hist(df['Revenue'].dropna(), bins=40, color='#3B82F6', alpha=0.7, edgecolor='black')
mean_rev = df['Revenue'].mean()
median_rev = df['Revenue'].median()
axes[0,0].axvline(mean_rev, color='red', linestyle='--', linewidth=2.5, label=f'Mean: {mean_rev:.2f}')
axes[0,0].axvline(median_rev, color='orange', linestyle='--', linewidth=2.5, label=f'Median: {median_rev:.2f}')
axes[0,0].set_title('Revenue Distribution', fontsize=14, fontweight='bold')
axes[0,0].set_xlabel('Revenue', fontsize=12)
axes[0,0].set_ylabel('Frequency', fontsize=12)
axes[0,0].legend(fontsize=10)
axes[0,0].grid(alpha=0.3)
textstr = f'Std: {df["Revenue"].std():.2f}\nSkew: {df["Revenue"].skew():.2f}'
props = dict(boxstyle='round', facecolor='wheat', alpha=0.8)
axes[0,0].text(0.72, 0.95, textstr, transform=axes[0,0].transAxes, fontsize=10,
               verticalalignment='top', bbox=props)

# Profit Margin
axes[0,1].hist(df['ProfitMargin'].dropna(), bins=40, color='#10B981', alpha=0.7, edgecolor='black')
mean_pm = df['ProfitMargin'].mean()
median_pm = df['ProfitMargin'].median()
axes[0,1].axvline(mean_pm, color='red', linestyle='--', linewidth=2.5, label=f'Mean: {mean_pm:.2f}')
axes[0,1].axvline(median_pm, color='orange', linestyle='--', linewidth=2.5, label=f'Median: {median_pm:.2f}')
axes[0,1].set_title('Profit Margin Distribution', fontsize=14, fontweight='bold')
axes[0,1].set_xlabel('Profit Margin (%)', fontsize=12)
axes[0,1].set_ylabel('Frequency', fontsize=12)
axes[0,1].legend(fontsize=10)
axes[0,1].grid(alpha=0.3)
textstr = f'Std: {df["ProfitMargin"].std():.2f}\nSkew: {df["ProfitMargin"].skew():.2f}'
axes[0,1].text(0.72, 0.95, textstr, transform=axes[0,1].transAxes, fontsize=10,
               verticalalignment='top', bbox=props)

# Market Cap
axes[1,0].hist(df['MarketCap'].dropna(), bins=40, color='#EF4444', alpha=0.7, edgecolor='black')
mean_mc = df['MarketCap'].mean()
median_mc = df['MarketCap'].median()
axes[1,0].axvline(mean_mc, color='red', linestyle='--', linewidth=2.5, label=f'Mean: {mean_mc:.2f}')
axes[1,0].axvline(median_mc, color='orange', linestyle='--', linewidth=2.5, label=f'Median: {median_mc:.2f}')
axes[1,0].set_title('Market Cap Distribution', fontsize=14, fontweight='bold')
axes[1,0].set_xlabel('Market Cap', fontsize=12)
axes[1,0].set_ylabel('Frequency', fontsize=12)
axes[1,0].legend(fontsize=10)
axes[1,0].grid(alpha=0.3)
textstr = f'Std: {df["MarketCap"].std():.2f}\nSkew: {df["MarketCap"].skew():.2f}'
axes[1,0].text(0.72, 0.95, textstr, transform=axes[1,0].transAxes, fontsize=10,
               verticalalignment='top', bbox=props)

# Growth Rate
axes[1,1].hist(df['GrowthRate'].dropna(), bins=40, color='#8B5CF6', alpha=0.7, edgecolor='black')
mean_gr = df['GrowthRate'].mean()
median_gr = df['GrowthRate'].median()
axes[1,1].axvline(mean_gr, color='red', linestyle='--', linewidth=2.5, label=f'Mean: {mean_gr:.2f}')
axes[1,1].axvline(median_gr, color='orange', linestyle='--', linewidth=2.5, label=f'Median: {median_gr:.2f}')
axes[1,1].axvline(0, color='black', linestyle='-', linewidth=1.5, alpha=0.5)
axes[1,1].set_title('Growth Rate Distribution', fontsize=14, fontweight='bold')
axes[1,1].set_xlabel('Growth Rate (%)', fontsize=12)
axes[1,1].set_ylabel('Frequency', fontsize=12)
axes[1,1].legend(fontsize=10)
axes[1,1].grid(alpha=0.3)
textstr = f'Std: {df["GrowthRate"].std():.2f}\nSkew: {df["GrowthRate"].skew():.2f}'
axes[1,1].text(0.72, 0.95, textstr, transform=axes[1,1].transAxes, fontsize=10,
               verticalalignment='top', bbox=props)

plt.tight_layout()
plt.savefig('../output/financial_distributions.png', dpi=300, bbox_inches='tight')
plt.show()

# 3.3 Environmental Metrics - Advanced Boxplots with Outlier Statistics
fig, axes = plt.subplots(1, 3, figsize=(20, 7))
fig.suptitle('Environmental Metrics - Outlier Detection & Analysis', fontsize=18, fontweight='bold')

env_cols = ['CarbonEmissions', 'WaterUsage', 'EnergyConsumption']
colors_env = ['#EF4444', '#3B82F6', '#F59E0B']

for idx, (col, color) in enumerate(zip(env_cols, colors_env)):
    # Create boxplot with custom styling
    bp = axes[idx].boxplot(df[col].dropna(), patch_artist=True, widths=0.5,
                           boxprops=dict(facecolor=color, alpha=0.7, linewidth=2),
                           medianprops=dict(color='black', linewidth=3),
                           whiskerprops=dict(color='black', linewidth=2),
                           capprops=dict(color='black', linewidth=2),
                           flierprops=dict(marker='o', markerfacecolor='darkred', 
                                          markersize=8, alpha=0.6, markeredgecolor='black'))
    
    axes[idx].set_title(col, fontsize=14, fontweight='bold')
    axes[idx].set_ylabel('Value', fontsize=12)
    axes[idx].grid(alpha=0.3, axis='y', linestyle='--')
    
    # Calculate detailed statistics
    Q1 = df[col].quantile(0.25)
    Q2 = df[col].quantile(0.50)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)][col]
    
    # Add statistics textbox
    textstr = f'Median: {Q2:.0f}\nIQR: {IQR:.0f}\nLower: {lower_bound:.0f}\nUpper: {upper_bound:.0f}\nOutliers: {len(outliers)} ({len(outliers)/len(df)*100:.1f}%)'
    props = dict(boxstyle='round', facecolor='lightgray', alpha=0.9)
    axes[idx].text(0.5, 0.98, textstr, transform=axes[idx].transAxes, fontsize=11,
                   verticalalignment='top', horizontalalignment='center', bbox=props)

plt.tight_layout()
plt.savefig('../output/environmental_boxplots.png', dpi=300, bbox_inches='tight')
plt.show()

# 3.4 Additional Boxplots - ESG Scores with Outlier Detection
fig, axes = plt.subplots(1, 4, figsize=(20, 6))
fig.suptitle('ESG Scores - Boxplot Outlier Detection', fontsize=18, fontweight='bold')

esg_cols = ['ESG_Overall', 'ESG_Environmental', 'ESG_Social', 'ESG_Governance']
colors_esg = ['#10B981', '#059669', '#FC8181', '#F59E0B']

for idx, (col, color) in enumerate(zip(esg_cols, colors_esg)):
    bp = axes[idx].boxplot(df[col].dropna(), patch_artist=True, widths=0.6,
                           boxprops=dict(facecolor=color, alpha=0.7),
                           medianprops=dict(color='darkred', linewidth=2),
                           whiskerprops=dict(color='black', linewidth=1.5),
                           capprops=dict(color='black', linewidth=1.5),
                           flierprops=dict(marker='o', markerfacecolor='red', markersize=8, alpha=0.5))
    
    axes[idx].set_title(col.replace('ESG_', ''), fontsize=14, fontweight='bold')
    axes[idx].set_ylabel('Score', fontsize=12)
    axes[idx].grid(alpha=0.3, axis='y')
    
    # Calculate IQR and outliers
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    outliers = df[(df[col] < Q1 - 1.5*IQR) | (df[col] > Q3 + 1.5*IQR)][col]
    
    # Add statistics text
    textstr = f'Q1: {Q1:.2f}\nQ3: {Q3:.2f}\nIQR: {IQR:.2f}\nOutliers: {len(outliers)}'
    props = dict(boxstyle='round', facecolor='lightblue', alpha=0.8)
    axes[idx].text(0.5, 0.02, textstr, transform=axes[idx].transAxes, fontsize=10,
                   verticalalignment='bottom', horizontalalignment='center', bbox=props)

plt.tight_layout()
plt.savefig('../output/esg_boxplots_outliers.png', dpi=300, bbox_inches='tight')
plt.show()

# 3.5 Additional Boxplots - Financial Metrics
# fig, axes = plt.subplots(1, 4, figsize=(20, 6))
# fig.suptitle('Financial Metrics - Boxplot Outlier Detection', fontsize=18, fontweight='bold')

# financial_cols = ['Revenue', 'ProfitMargin', 'MarketCap', 'GrowthRate']
# colors_fin = ['#3B82F6', '#10B981', '#EF4444', '#8B5CF6']

# for idx, (col, color) in enumerate(zip(financial_cols, colors_fin)):
#     bp = axes[idx].boxplot(df[col].dropna(), patch_artist=True, widths=0.6,
#                            boxprops=dict(facecolor=color, alpha=0.7),
#                            medianprops=dict(color='darkred', linewidth=2),
#                            whiskerprops=dict(color='black', linewidth=1.5),
#                            capprops=dict(color='black', linewidth=1.5),
#                            flierprops=dict(marker='D', markerfacecolor='red', markersize=8, alpha=0.5))
    
#     axes[idx].set_title(col, fontsize=14, fontweight='bold')
#     axes[idx].set_ylabel('Value', fontsize=12)
#     axes[idx].grid(alpha=0.3, axis='y')
    
#     # Calculate IQR and outliers
#     Q1 = df[col].quantile(0.25)
#     Q3 = df[col].quantile(0.75)
#     IQR = Q3 - Q1
#     outliers = df[(df[col] < Q1 - 1.5*IQR) | (df[col] > Q3 + 1.5*IQR)][col]
    
#     textstr = f'Q1: {Q1:.2f}\nQ3: {Q3:.2f}\nIQR: {IQR:.2f}\nOutliers: {len(outliers)}'
#     props = dict(boxstyle='round', facecolor='lightyellow', alpha=0.8)
#     axes[idx].text(0.5, 0.02, textstr, transform=axes[idx].transAxes, fontsize=10,
#                    verticalalignment='bottom', horizontalalignment='center', bbox=props)
    

# Scatter Plots - Key Relationships
print("\n🔍 Creating scatter plots for key relationships...")

# Define key relationships to explore
relationships = [
    ('ESG_Overall', 'Revenue'),
    ('ESG_Overall', 'ProfitMargin'),
    ('ESG_Overall', 'CarbonEmissions'),
    ('ESG_Environmental', 'CarbonEmissions'),
    ('ESG_Environmental', 'WaterUsage'),
    ('ESG_Environmental', 'EnergyConsumption'),
    ('CarbonEmissions', 'WaterUsage'),
    ('CarbonEmissions', 'EnergyConsumption'),
    ('WaterUsage', 'EnergyConsumption'),
    ('Revenue', 'MarketCap'),
    ('Revenue', 'ProfitMargin'),
    ('GrowthRate', 'ESG_Overall'),
    ('MarketCap', 'ESG_Overall'),
    ('ProfitMargin', 'GrowthRate'),
    ('ESG_Social', 'ESG_Governance'),
    ('ESG_Environmental', 'ESG_Social')
]

# Define a color palette for scatter plots
import seaborn as sns
colors = sns.color_palette('tab20', len(relationships))

n_relationships = len(relationships)
n_rows_scatter = (n_relationships + 3) // 4
fig, axes = plt.subplots(n_rows_scatter, 4, figsize=(20, n_rows_scatter * 4))
axes = axes.flatten()

fig.suptitle('Scatter Plots - Bivariate Relationships', fontsize=20, fontweight='bold')

for idx, (col1, col2) in enumerate(relationships):
    color = colors[idx % len(colors)]
    
    # Scatter plot
    axes[idx].scatter(df[col1], df[col2], alpha=0.5, s=30, c=[color], edgecolors='black', linewidth=0.5)
    
    # Calculate and add correlation
    correlation = df[col1].corr(df[col2])
    
    # Add trend line
    valid = df[[col1, col2]].dropna()
    z = np.polyfit(valid[col1], valid[col2], 1)
    p = np.poly1d(z)
    x_line = np.linspace(valid[col1].min(), valid[col1].max(), 100)
    axes[idx].plot(x_line, p(x_line), "r--", linewidth=2, alpha=0.8)
    
    axes[idx].set_xlabel(col1, fontsize=10, fontweight='bold')
    axes[idx].set_ylabel(col2, fontsize=10, fontweight='bold')
    axes[idx].set_title(f'{col1} vs {col2}', fontsize=11, fontweight='bold')
    axes[idx].grid(alpha=0.3)
    
    # Add correlation textbox
    textstr = f'Correlation: {correlation:.3f}\n'
    if abs(correlation) > 0.7:
        textstr += 'Strong'
    elif abs(correlation) > 0.4:
        textstr += 'Moderate'
    else:
        textstr += 'Weak'
    
    props = dict(boxstyle='round', facecolor='yellow' if abs(correlation) > 0.5 else 'lightgray', alpha=0.8)
    axes[idx].text(0.05, 0.95, textstr, transform=axes[idx].transAxes, fontsize=9,
                   verticalalignment='top', bbox=props)

# Hide extra subplots
for idx in range(n_relationships, len(axes)):
    axes[idx].set_visible(False)

plt.tight_layout()
plt.savefig('../output/all_scatter_plots.png', dpi=300, bbox_inches='tight')
plt.show()
print("✅ Scatter plots created and saved")

print("\n" + "="*80)
print("5. MULTIVARIATE ANALYSIS - CORRELATION HEATMAPS")
print("="*80)

# 5.1 Overall Correlation Heatmap
print("\n🔥 Creating correlation heatmap for all features...")
# Define numeric columns
numeric_cols = df.select_dtypes(include=['float64', 'int64']).columns.tolist()
plt.figure(figsize=(16, 14))
correlation_matrix = df[numeric_cols].corr()

# Create heatmap
mask = np.triu(np.ones_like(correlation_matrix, dtype=bool), k=1)
sns.heatmap(correlation_matrix, annot=True, fmt='.2f', cmap='RdYlGn', center=0,
            square=True, linewidths=1, cbar_kws={"shrink": 0.8}, 
            vmin=-1, vmax=1, mask=None)

plt.title('Correlation Heatmap - All Numerical Features', fontsize=18, fontweight='bold', pad=20)
plt.tight_layout()
plt.savefig('../output/correlation_heatmap_full.png', dpi=300, bbox_inches='tight')
plt.show()
print("✅ Full correlation heatmap created")

# 5.2 ESG Components Correlation
print("\n🎯 Creating ESG-specific correlation heatmap...")
esg_cols = ['ESG_Overall', 'ESG_Environmental', 'ESG_Social', 'ESG_Governance']
esg_corr = df[esg_cols].corr()

plt.figure(figsize=(10, 8))
sns.heatmap(esg_corr, annot=True, fmt='.3f', cmap='Greens', center=0,
            square=True, linewidths=2, cbar_kws={"shrink": 0.8}, vmin=0, vmax=1)
plt.title('ESG Components - Correlation Matrix', fontsize=16, fontweight='bold', pad=15)
plt.tight_layout()
plt.savefig('../output/correlation_esg_components.png', dpi=300, bbox_inches='tight')
plt.show()
print("✅ ESG correlation heatmap created")

# 5.3 Financial Metrics Correlation
print("\n💰 Creating financial metrics correlation heatmap...")
financial_cols = ['Revenue', 'ProfitMargin', 'MarketCap', 'GrowthRate']
financial_corr = df[financial_cols].corr()

plt.figure(figsize=(10, 8))
sns.heatmap(financial_corr, annot=True, fmt='.3f', cmap='Blues', center=0,
            square=True, linewidths=2, cbar_kws={"shrink": 0.8}, vmin=-1, vmax=1)
plt.title('Financial Metrics - Correlation Matrix', fontsize=16, fontweight='bold', pad=15)
plt.tight_layout()
plt.savefig('../output/correlation_financial_metrics.png', dpi=300, bbox_inches='tight')
plt.show()
print("✅ Financial correlation heatmap created")

# 5.4 Environmental Metrics Correlation
print("\n🌍 Creating environmental metrics correlation heatmap...")
env_cols = ['CarbonEmissions', 'WaterUsage', 'EnergyConsumption']
env_corr = df[env_cols].corr()

plt.figure(figsize=(8, 6))
sns.heatmap(env_corr, annot=True, fmt='.3f', cmap='Reds', center=0,
            square=True, linewidths=2, cbar_kws={"shrink": 0.8}, vmin=0, vmax=1)
plt.title('Environmental Metrics - Correlation Matrix', fontsize=16, fontweight='bold', pad=15)
plt.tight_layout()
plt.savefig('../output/correlation_environmental_metrics.png', dpi=300, bbox_inches='tight')
plt.show()
print("✅ Environmental correlation heatmap created")

# 5.5 ESG vs Financial Performance Correlation
print("\n📊 Creating ESG vs Financial correlation heatmap...")
esg_financial_cols = esg_cols + financial_cols
esg_financial_corr = df[esg_financial_cols].corr()

plt.figure(figsize=(12, 10))
sns.heatmap(esg_financial_corr, annot=True, fmt='.2f', cmap='coolwarm', center=0,
            square=True, linewidths=1, cbar_kws={"shrink": 0.8}, vmin=-1, vmax=1)
plt.title('ESG vs Financial Performance - Correlation Matrix', fontsize=16, fontweight='bold', pad=15)
plt.tight_layout()
plt.savefig('../output/correlation_esg_financial.png', dpi=300, bbox_inches='tight')
plt.show()
print("✅ ESG vs Financial correlation heatmap created")

# 5.6 Multicollinearity Check - Variance Inflation Factor (VIF)
print("\n" + "-"*80)
print("MULTICOLLINEARITY ANALYSIS - VIF (Variance Inflation Factor)")
print("-"*80)

from scipy.stats import zscore

# Select features for VIF analysis (exclude highly correlated features manually if needed)
vif_features = ['ESG_Overall', 'Revenue', 'ProfitMargin', 'MarketCap', 'GrowthRate', 
                'CarbonEmissions', 'WaterUsage', 'EnergyConsumption']

# Simple VIF calculation
vif_data = pd.DataFrame()
vif_data["Feature"] = vif_features

# Calculate VIF for each feature
from sklearn.linear_model import LinearRegression

vif_values = []
for i, feature in enumerate(vif_features):
    X = df[vif_features].drop(columns=[feature])
    y = df[feature]
    # Drop rows with NaN in either X or y
    valid_idx = X.notnull().all(axis=1) & y.notnull()
    X_valid = X.loc[valid_idx]
    y_valid = y.loc[valid_idx]
    
    if len(X_valid) > 0:
        model = LinearRegression()
        model.fit(X_valid, y_valid)
        r_squared = model.score(X_valid, y_valid)
        vif = 1 / (1 - r_squared) if r_squared < 0.9999 else float('inf')
        vif_values.append(vif)
    else:
        vif_values.append(0)

vif_data["VIF"] = vif_values
vif_data = vif_data.sort_values('VIF', ascending=False)

print(vif_data)
print("\nVIF Interpretation:")
print("  VIF < 5: Low multicollinearity")
print("  5 < VIF < 10: Moderate multicollinearity")
print("  VIF > 10: High multicollinearity (problematic for regression)")

# Visualize VIF
fig, ax = plt.subplots(figsize=(12, 6))
colors_vif = ['green' if v < 5 else 'orange' if v < 10 else 'red' for v in vif_data['VIF']]
bars = ax.barh(vif_data['Feature'], vif_data['VIF'], color=colors_vif, alpha=0.7, edgecolor='black', linewidth=1.5)
ax.axvline(x=5, color='orange', linestyle='--', linewidth=2, label='Moderate threshold (VIF=5)')
ax.axvline(x=10, color='red', linestyle='--', linewidth=2, label='High threshold (VIF=10)')
ax.set_xlabel('VIF Value', fontsize=12, fontweight='bold')
ax.set_title('Variance Inflation Factor (VIF) - Multicollinearity Check', fontsize=16, fontweight='bold')
ax.legend(fontsize=10)
ax.grid(alpha=0.3, axis='x')

# Add value labels
for i, (feature, vif) in enumerate(zip(vif_data['Feature'], vif_data['VIF'])):
    ax.text(vif + 0.5, i, f'{vif:.2f}', va='center', fontweight='bold')

plt.tight_layout()
plt.savefig('../output/multicollinearity_vif.png', dpi=300, bbox_inches='tight')
plt.show()
print("✅ VIF analysis completed")

# 5.7 Pair Plot for Selected Features (subset to avoid overcrowding)
print("\n🎨 Creating pair plot for key features...")
key_features = ['ESG_Overall', 'Revenue', 'CarbonEmissions', 'ProfitMargin', 'GrowthRate']

# Sample data if too large
if len(df) > 1000:
    df_sample = df[key_features].sample(1000, random_state=42)
else:
    df_sample = df[key_features]

pair_plot = sns.pairplot(df_sample, diag_kind='kde', plot_kws={'alpha': 0.6, 's': 30, 'edgecolor': 'black'},
                         diag_kws={'color': '#10B981', 'linewidth': 2})
pair_plot.fig.suptitle('Pair Plot - Key Features Relationships', y=1.01, fontsize=18, fontweight='bold')
plt.tight_layout()
plt.savefig('../output/pairplot_key_features.png', dpi=300, bbox_inches='tight')
plt.show()
print("✅ Pair plot created")



plt.tight_layout()
plt.savefig('../output/financial_boxplots_outliers.png', dpi=300, bbox_inches='tight')
plt.show()

print("\n✅ Univariate Analysis Complete - All visualizations saved to 'outputs' folder")

```
